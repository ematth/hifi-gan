```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
```

```python
def plotting(file: str, fig: plt.figure = plt.figure()) -> plt.plot:
    file = file.replace('.wav', '', 1); file = file.replace('test_files/', '', 1)
    sr, x = read(f'test_files/{file}.wav')
    #print(sr, x)
    plt.plot(x)
    return fig

f = plt.figure()
f = plotting('test_files/minuet_48k.wav', fig=f)
#plotting('minuet')
```
