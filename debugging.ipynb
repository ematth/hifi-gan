```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
import pandas as pd
```

```python
def plotting(file: str, fig: plt.figure = plt.figure()) -> tuple[plt.plot, int, np.ndarray]:
    file = file.replace('.wav', '', 1); file = file.replace('test_files/', '', 1)
    sr, x = read(f'test_files/{file}.wav')
    #print(sr, x)
    plt.plot(x)
    return fig, sr, x

a = plt.figure()
a, sr, x = plotting('test_files/minuet_48k.wav', fig=a)
#plotting('minuet')
```

```python
b = plt.figure()
b = plotting('test_files/minuet.wav', fig=b)
```
